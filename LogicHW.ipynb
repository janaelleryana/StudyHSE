```python
user_role = "admin"
has_api_key = True
model_available = True
request_count_today = 95
daily_limit = 100

# Условия доступа
has_access = ((user_role == "admin" or has_api_key) and
              model_available and
              request_count_today < daily_limit)

print("Доступ разрешен:", has_access)
```

Задание 2

```python
text = "Привет! Как дела?"
banned_words = ["мат", "оскорбление", "спам", "реклама"]

text_lower = text.lower()
contains_banned_words = any(word in text_lower for word in banned_words)
is_english_only = all(ord(c) < 128 for c in text if c != ' ')

is_valid = (1 <= len(text) <= 1000 and
            not contains_banned_words and
            not is_english_only)

print("Текст прошел проверку:", is_valid)
```

Задание 3

```python
user_permissions = ["read", "analyze", "moderate"]
content_categories = ["text", "image"]
current_content_type = "text"

can_moderate = ("moderate" in user_permissions and
                current_content_type in content_categories)

can_analyze = ("analyze" in user_permissions and
               len(content_categories) > 0)

print("Может модерировать:", can_moderate)
print("Может анализировать:", can_analyze)
```

Задание 4

```python
reviews_sentiments = [0.8, 0.6, -0.3, 0.9, 0.7]

all_positive = min(reviews_sentiments) > 0

has_negative = min(reviews_sentiments) < 0

has_very_negative = min(reviews_sentiments) < -0.5

total_reviews = len(reviews_sentiments)
positive_count = sum(1 for _ in filter(lambda x: x > 0, reviews_sentiments))
majority_positive = positive_count > total_reviews / 2

print("Все отзывы позитивные:", all_positive)
print("Есть негативные:", has_negative)
print("Есть очень негативные:", has_very_negative)
print("Большинство позитивные:", majority_positive)
```

Задание 5

```python
user_tier = "premium"
subscription_active = True
available_models = ["gpt-3", "bert", "word2vec", "fasttext"]
premium_models = ["gpt-3", "bert", "gpt-4"]
user_quota = {"used": 45, "total": 100}

has_any_premium_model = any(model in available_models for model in premium_models)

has_premium_access = (user_tier == "premium" and
                      subscription_active and
                      has_any_premium_model and
                      user_quota["used"] < user_quota["total"])

has_basic_access = (len(available_models) > 0 and
                    user_quota["used"] < user_quota["total"])

print("Доступ к премиум-моделям:", has_premium_access)
print("Доступ к базовым функциям:", has_basic_access)
```
